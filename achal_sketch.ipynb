```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
image = cv2.imread("C:\\Users\\ankit\\OneDrive\\Desktop\\t2\\sai_pallavi.jpg")
```

```python
plt.imshow(image)
```

```python
grey_filter = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("output1.jpg", grey_filter)
```

```python
plt.imshow(grey_filter)
```

```python
invert = cv2.bitwise_not(grey_filter)
```

```python
cv2.imwrite("output2.jpg", invert)
```

```python
plt.imshow(invert)
```

```python
blur = cv2.GaussianBlur(invert, (21,21),0)
```

```python
cv2.imwrite("output3.jpg", blur)
```

```python
plt.imshow(blur)
```

```python
invertedblur = cv2.bitwise_not(blur)
```

```python
cv2.imwrite("output4.jpg", invertedblur)
```

```python
plt.imshow(invertedblur)
```

```python
sketch_filter = cv2.divide(grey_filter, invertedblur, scale=256.0)
```

```python
cv2.imwrite("output.png", sketch_filter)
```

```python
plt.imshow(sketch_filter)
```
